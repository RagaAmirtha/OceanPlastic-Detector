```python
import os
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
import os

dataset_path = r"D:\dataset\raw_data"
train_path = os.path.join(dataset_path, "train")
```

```python
print("Dataset Path:", dataset_path)
print("Train Path:", train_path)
print("Folders inside train:", os.listdir(train_path))
```

```python
classes = os.listdir(train_path)

for cls in classes:
    class_folder = os.path.join(train_path, cls)
    print(f"{cls}: {len(os.listdir(class_folder))} images")
```

```python
import matplotlib.pyplot as plt
from PIL import Image

plt.figure(figsize=(12, 6))

for i, cls in enumerate(classes):
    class_folder = os.path.join(train_path, cls)
    img_name = os.listdir(class_folder)[0]
    img_path = os.path.join(class_folder, img_name)

    img = Image.open(img_path)

    plt.subplot(1, len(classes), i + 1)
    plt.imshow(img)
    plt.title(cls)
    plt.axis("off")

plt.show()
```

```python
counts = []

for cls in classes:
    class_folder = os.path.join(train_path, cls)
    counts.append(len(os.listdir(class_folder)))

plt.figure(figsize=(6, 4))
plt.bar(classes, counts)
plt.title("Class Distribution")
plt.xlabel("Classes")
plt.ylabel("Number of Images")
plt.show()
```

```python
sizes = []

for cls in classes:
    class_folder = os.path.join(train_path, cls)

    for img_name in os.listdir(class_folder)[:20]:
        img_path = os.path.join(class_folder, img_name)
        img = Image.open(img_path)
        sizes.append(img.size)

print("Sample image sizes:")
print(sizes[:10])
```

## Observation: Class Distribution
The dataset is nearly balanced, with approximately 880 no-plastic images and 840 plastic images.
This balanced distribution is beneficial for training a CNN model because it reduces prediction bias.
